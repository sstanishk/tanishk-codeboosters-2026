```python
cs_students = df[df['department']== 'Computer Science']
print(f"Number of Computer Science students: {len(cs_students)}")
print("\nComputer Science students:")
print(cs_students[['name','math_score','programming_score','attendance_percentage']])
```

```python
high_attendance_students = df[df['attendance_percentage'] > 90]
print(f"Number of students with attendance above 90%: {len(high_attendance_students)}")
print("\nStudents with attendance above 90%:")
print(high_attendance_students[['name', 'attendance_percentage', 'department']])
```

```python
df['average_score'] = (df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score'] )/ 4

top_students = df[df['average_score'] > 90] # Filter by average score > 90

print(f"Number of students with an average score above 90: {len(top_students)}")

print("\nStudents with high average scores:")
print(top_students[['name', 'department', 'average_score']])
```

```python
if 'average_score' not in df.columns:
    df['average_score'] = (df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']) / 4

# Calculate average score per department
average_score_per_department = df.groupby('department')['average_score'].mean().reset_index()
average_score_per_department = average_score_per_department.rename(columns={'average_score': 'department_average_score'})

print("Average score per department:")
print(average_score_per_department.round(2))
```
