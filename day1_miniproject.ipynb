```python
#STEP 1:Import Libraries
import pandas as pd
print("Libraries imported successfully!")
```

```python
df = pd.read_csv('student_performance.csv')
```

```python
print("="*35)
print("STUDENT PERFORMANCE EXPLORER REPORT")
print("="*35)
print("\n[1] DATASET OVERVIEW")
print("-"*35)
print("Total Students:", df.shape[0])
print("Total Columns:", df.shape[1])
print("Departments:", len(df['department'].unique()))
```

```python
print("\n[2] DEPARTMENT-WISE STUDENT COUNT")
print("-"*35)
print(df['department'].value_counts())
```

```python
print("\n[3] SCORE ANALYSIS")
print("-"*35)
print("Average math Score:", df['math_score'].mean().round(2))
print("Minimum math Score:", df['math_score'].min())
print("Maximum math Score:", df['math_score'].max())
print("Average programming Score:", df['programming_score'].mean().round(2))
print("Average attendance  percentage:", df['attendance_percentage'].mean().round(2))



```

```python
print("\n[4] GENDER WISE AVERAGE  SCORE")
print("-"*35)
total_scores = df[['math_score', 'programming_score','english_score','science_score']].sum(axis=1)
df['total_score'] = total_scores
gender_scores = df.groupby('gender')['total_score'].mean().round(2)
print(gender_scores)
```

```python
print("\n[5] TOP 5 PERFORMERS")
print("_"*40)
top5 = df.sort_values('total_score', ascending=False).head(5)
for index, row in top5.iterrows():
    print(f" {row['name']:<20} | {row['department']:<18} | Score:{row['total_score']}")

```

```python
print("\n[6] STUDENTS WITH LOW ATTENDANCE")
print("_"*55)
low_attendance = df[df['attendance_percentage'] < 75]
for index, row in low_attendance.iterrows():
    print(f" {row['name']:<20} | {row['department']:<18} | Attendance:{row['attendance_percentage']}")
```
